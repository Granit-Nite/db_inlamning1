```python
import pandas as pd

college = pd.read_csv("../DB_INLAMNING1/College.csv") 
```

```python
### OBS dessa kommentarer är för min egen del för att jag skall lära mig vad de olika koderna gör


# college #- börjar med index 0,1,2, samt innehåller kolumnen "Unnamed:"

# college2 = pd.read_csv('College.csv', index_col=0)
# college2  # tar bort indexen i början, dvs börjar på namnet istället, samt tar bort kolumnen "Unnamed"

college3 = college.rename({'Unnamed: 0': 'College'}, axis = 1)
college3 # Vi får tillbaka indexen samt "Unnamed" heter nu "College"

college3 = college3.set_index('College')
college3 #kolumnen College blir private
```

```python
college = college3
college.describe()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame(college) 


C = ['Top10perc', 'Apps', 'Enroll']
A = df

pd.plotting.scatter_matrix(A[C])


plt.show()
```
